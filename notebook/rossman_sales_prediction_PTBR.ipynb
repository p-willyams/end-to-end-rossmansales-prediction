### 4.3 Funções Adicionais

Nesta etapa, são importadas as **funções auxiliares** que serão utilizadas ao longo de todo o notebook.  
Essas funções foram criadas para **otimizar o fluxo de trabalho**, tornando o código mais **organizado, reutilizável e legível**.


```python
def snake_case(lst):
    """
    Recebe uma lista de strings e retorna uma nova lista no formato snake_case.
    """
    def convert(s):
        s = s.replace(' ', '_')
        new_s = ""
        for i, c in enumerate(s):
            if c.isupper():
                if i > 0 and (s[i-1].islower() or (i+1 < len(s) and s[i+1].islower())):
                    new_s += "_"
                new_s += c.lower()
            else:
                new_s += c
        return new_s

    return [convert(s) for s in lst]

# Avaliação do MPE (Mean Percentage Error) do modelo
def mean_percentage_error(y_true, y_pred):
    """Calcula o erro percentual médio (MPE)."""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean((y_true - y_pred) / y_true) * 100


def cramers_v_bias_corrected(confusion_matrix):
    """Calcula o V de Cramer (correto para bias) dado uma tabela de contingência."""
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    phi2 = chi2 / n
    # bias correction
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    denominator = min((kcorr-1), (rcorr-1))
    if denominator == 0:
        return np.nan
    return np.sqrt(phi2corr / denominator)

def calcular_metricas(y_true, y_pred, nome_modelo="modelo"):
    """
    Calcula e retorna as principais métricas de avaliação de modelos de regressão em um DataFrame.
    """
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = r2_score(y_true, y_pred)
    df_metricas = pd.DataFrame({
        'modelo': [nome_modelo],
        'rmse': [rmse],
        'mae': [mae],
        'mape': [mape],
        'r2': [r2]
    })
    return df_metricas

def cross_validation(x_training, kfold, model_name, model, verbose=False):
    """
    Executa cross-validation usando método de validation por janela de tempo.
    Assumindo intervalos fixos de 6 semanas como janelas de validação.

    Parâmetros:
        x_training (DataFrame): dados de treino, deve conter as colunas 'date' e 'sales'.
        kfold (int): número de folds (partições de janela de tempo, ex: 5).
        model_name (str): nome do modelo.
        model: objeto modelo sklearn (ou similar), já instanciado.
        verbose (bool): se True, imprime informações do fold.

    Retorno:
        DataFrame com médias/STD de MAE, MAPE, RMSE e R2 para todos os folds.
    """
    mae_list = []
    mape_list = []
    rmse_list = []
    r2_list = []

    window_days = 6 * 7
    if not np.issubdtype(x_training['date'].dtype, np.datetime64):
        x_training = x_training.copy()
        x_training['date'] = pd.to_datetime(x_training['date'])

    for k in reversed(range(1, kfold+1)):
        if verbose:
            print('\nKFold Number: {}'.format(k))
        max_date = x_training['date'].max()
        validation_start_date = max_date - pd.Timedelta(days=k*window_days)
        validation_end_date = max_date - pd.Timedelta(days=(k-1)*window_days)
        
        training = x_training[x_training['date'] < validation_start_date]
        validation = x_training[(x_training['date'] >= validation_start_date) & (x_training['date'] <= validation_end_date)]
        
        if len(training) == 0 or len(validation) == 0:
            if verbose:
                print("Skipping fold {} (not enough data)".format(k))
            continue

        xtraining = training.drop(['date', 'sales'], axis=1)
        ytraining = training['sales']
        xvalidation = validation.drop(['date', 'sales'], axis=1)
        yvalidation = validation['sales']


        m = model.fit(xtraining, ytraining)
        yhat = m.predict(xvalidation)

        y_true = np.expm1(yvalidation)
        y_pred = np.expm1(yhat)

        mae_value = mean_absolute_error(y_true, y_pred)
        mape_value = np.mean(np.abs((y_true - y_pred) / np.where(y_true == 0, 1, y_true))) * 100
        mse_value = mean_squared_error(y_true, y_pred)
        rmse_value = np.sqrt(mse_value)
        r2_value = r2_score(y_true, y_pred)

        mae_list.append(mae_value)
        mape_list.append(mape_value)
        rmse_list.append(rmse_value)
        r2_list.append(r2_value)
    df_result = pd.DataFrame({
        'Model Name': [model_name],
        'MAE CV': [str(np.round(np.mean(mae_list), 2)) + ' +/- ' + str(np.round(np.std(mae_list), 2))],
        'MAPE CV': [str(np.round(np.mean(mape_list), 2)) + ' +/- ' + str(np.round(np.std(mape_list), 2))],
        'RMSE CV': [str(np.round(np.mean(rmse_list), 2)) + ' +/- ' + str(np.round(np.std(rmse_list), 2))],
        'R2 CV': [str(np.round(np.mean(r2_list), 4)) + ' +/- ' + str(np.round(np.std(r2_list), 4))]
    })
    return df_result
```

#### 5.4.1 Competition Distance

Dessa forma, é razoável **atribuir um valor alto** aos casos de `NaN`, permitindo que o modelo interprete essas situações como **ausência de competição relevante**. Iremos atribuir como `NaN` sendo 200.000

#### 5.4.2 Open Since

Nas colunas `competition_open_since_month` e `competition_open_since_year` indica **quando o concorrente mais próximo foi inaugurado**.  
Valores `NaN` podem ocorrer por duas razões principais:  
1. **Não há concorrente próximo**.  
2. **Não se sabe a data de abertura do concorrente**, seja porque ocorreu há muito tempo ou porque a loja em questão abriu depois do concorrente já existente.

Essas variáveis são importantes para o modelo, pois podem influenciar o comportamento das vendas ao refletir **tempo de competição local**.  

Como os modelos de *machine learning* não lidam bem com valores ausentes, é necessário substituir os `NaN`.  
A decisão tomada foi **preencher os valores com a data exata da venda**, ou seja, considerar que **o concorrente "começou a competir" no dia da observação**.  

Essa estratégia permite que o modelo interprete os `NaN` como **situações em que a competição ainda não exercia influência**, mantendo a consistência da variável sem distorcer os padrões de vendas.  
Se deixássemos `NaN` ou preenchêssemos de forma arbitrária, poderíamos **introduzir viés ou perder informações importantes** sobre o impacto da competição.

Iremos fazer o mesmo para `promo2_open_since_week` e `promo2_open_since_year`, seguindo a mesma lógica.

```python
#competition_open_since_month
df1['competition_open_since_month'] = df1['competition_open_since_month'].fillna(df1['date'].dt.month)

#competition_open_since_year
df1['competition_open_since_year'] = df1['competition_open_since_year'].fillna(df1['date'].dt.year)         

#promo2_since_week
df1['promo2_since_week'] = df1['promo2_since_week'].fillna(df1['date'].dt.isocalendar().week)

#promo2_since_year  
df1['promo2_since_year'] = df1['promo2_since_year'].fillna(df1['date'].dt.year) 

df1['competition_open_since_month'] = df1['competition_open_since_month'].astype(int)
df1['competition_open_since_year'] = df1['competition_open_since_year'].astype(int)
df1['promo2_since_week'] = df1['promo2_since_week'].astype(int)
df1['promo2_since_year'] = df1['promo2_since_year'].astype(int)
```

#### 5.5.1 Numerical Attributes

A análise descritiva de variáveis númericas é composta por dois grupos principais de medidas:

Medidas de Tendência Central:

Essas medidas resumem a variável em um único valor representativo:
- **Média (mean):** representa o valor médio da variável, sensível a outliers.  
- **Mediana (median):** valor central da distribuição; menos afetada por valores extremos e útil em distribuições assimétricas.  
- **Moda (mode):** valor que mais se repete, comum em variáveis categóricas.

Medidas de Dispersão:

Essas medidas indicam o quanto os dados variam em torno da média:
- **Desvio padrão (std):** indica o grau médio de dispersão em relação à média; quanto maior, mais variáveis são os dados.  
- **Mínimo (min):** menor valor observado.  
- **Máximo (max):** maior valor observado.  
- **Amplitude (range):** diferença entre o valor máximo e mínimo, representando a extensão total dos dados.  
- **Skewness (assimetria):** mostra se a distribuição é inclinada para a direita (positiva) ou esquerda (negativa).  
- **Kurtosis (curtose):** mede o grau de concentração dos dados nas caudas; valores altos indicam presença de outliers ou picos acentuados.

## 6. Feature Engineering

**Feature Engineering** refere-se ao processo de **criação de novas variáveis** a partir de outras já existentes no conjunto de dados, além da **seleção e filtragem** das variáveis e linhas que serão realmente relevantes para o estudo e para o treinamento do modelo.

Mas por que realizar essa etapa **antes da Análise Exploratória de Dados (EDA)**?

Assim, o Feature Engineering serve como uma preparação estruturada dos dados, garantindo que a etapa de Análise Exploratória ocorra sobre um conjunto limpo, coerente e mais rico em informações relevantes.

### 6.1 Criação de Hipóteses

Antes de tudo, precisamos de um **norte** para derivar as variáveis.  
Ou seja, é necessário definir **quais hipóteses serão validadas**, e a partir delas, identificar **quais variáveis são necessárias** para realizar essa validação.  

Seguir esse caminho torna a **Análise Exploratória de Dados (EDA)** mais **direcionada, eficiente e significativa**, pois cada transformação ou gráfico criado passa a ter um **propósito claro**, confirmar, ou refutar uma hipótese.  

Dessa forma, evitamos análises aleatórias e garantimos que os **insights obtidos sejam relevantes** para o entendimento do problema e para a construção de um modelo mais assertivo.  

#### 6.1.1 Mas como criar Hipóteses?

Antes de criar uma hipótese, precisamos identificar os **principais componentes que a moldam**, respondendo a algumas perguntas essenciais:  

**O Fenômeno:**  
Qual fenômeno estou modelando?  
> Neste caso, as **vendas**.  

**Agentes:**  
Quem são os agentes que atuam sobre esse fenômeno?  
> Neste caso, as **variáveis** do conjunto de dados.  

As **hipóteses** nada mais são do que **pensamentos estruturados sobre o fenômeno**.  
Para criá-las, é preciso compreender **como os agentes influenciam o fenômeno**, gerando uma lista de possíveis relações e comportamentos que poderão ser testados posteriormente.  

Essa **lista de hipóteses** pode ser construída de diversas formas:  
- A partir do **conhecimento de negócio** (experiência ou literatura sobre o setor);  
- Por meio de **reuniões com áreas específicas da empresa** (como marketing, operações ou vendas);  
- Ou até mesmo pela **curiosidade e senso comum** sobre o comportamento dos dados.  

Por exemplo:  
> “Lojas com mais clientes deveriam vender mais.”  

Essa é uma hipótese simples, mas direta, que pode, ou não, ser validada através dos dados.  


#### 6.1.2 Criação de Hipoteses

As **hipóteses** são a base da geração de **insights**, e esses insights podem surgir de duas formas principais:  

1. **A partir da surpresa**, quando os dados revelam algo inesperado;  
2. **Pela contraposição de crenças**, quando descobrimos verdades que desafiam o senso comum.  

Por exemplo:  
> “Lojas com mais clientes deveriam vender mais.”  

Se ao analisarmos os dados descobrimos que **essas lojas vendem menos**, isso nos gera um **insight poderoso**, pois quebra uma expectativa lógica.  

Durante o processo de análise, buscaremos hipóteses com esse potencial de **surpreender e revelar padrões ocultos**.  
No entanto, também é importante **validar hipóteses óbvias**, pois, embora pareçam certas, **elas podem ser invalidadas pelos dados**.  

Assim, a criação de hipóteses não é apenas uma etapa inicial, mas um exercício contínuo de **curiosidade, questionamento e verificação**.

Essas novas variáveis serão criadas de acordo com as **necessidades das hipóteses formuladas** e também com as **demandas específicas do modelo de Machine Learning** que será desenvolvido ao final deste notebook.

#### 6.2.1 Dias, Meses, Anos e Semanas do ano.

Muitos modelos de machine learning não lidam diretamente com dados no formato `datetime`. Por isso, é necessário derivar novas variáveis a partir da data original, que capturem aspectos importantes do fenômeno temporal. Neste caso, criamos três variáveis adicionais: **ano**, **mês** e **dia** e **Semanas do Ano**, que permitem ao modelo identificar padrões sazonais, mensais ou diários nas vendas.


```python
# Criar coluna 'year' como numérica
df2['year'] = df2['date'].dt.year.astype(int)

# Criar coluna 'month' como numérica
df2['month'] = df2['date'].dt.month.astype(int)

# Criar coluna 'day' como numérica
df2['day'] = df2['date'].dt.day.astype(int)

# Criar coluna 'week_of_year' como numérica
df2['week_of_year'] = df2['date'].dt.isocalendar().week.astype(int)
```

#### 6.2.2 Competition time month e Promo time week

- Hipótese 1: Lojas com competidores há mais tempo no mercado deveriam apresentar maiores vendas.  
- Hipótese 2: Lojas com promoções ativas por mais tempo deveriam apresentar maiores vendas.  

Para validar essas hipóteses, precisamos de variáveis que quantifiquem a duração da competição e da promoção. A partir das colunas `competition_open_since_year` e `competition_open_since_month`, e `promo_since_year` e `promo_since_week`, derivamos novas variáveis:  

- **competition_time_month**: mede, em meses, o tempo desde que a competição começou.  
- **promo_time_week**: mede, em semanas, o tempo de duração da promoção.  

Essas variáveis permitem avaliar se a presença prolongada de competidores ou promoções tem efeito sobre o volume de vendas das lojas.


```python
df2['competition_since'] = pd.to_datetime(
    dict(
        year=df2['competition_open_since_year'],
        month=df2['competition_open_since_month'],
        day=1
    )
)
df2['competition_time_month'] = ((df2['date'] - df2['competition_since']).dt.days / 30).round().astype(int)

# promo since
df2['promo_since'] = df2['promo2_since_year'].astype(int).astype(str) + '-' + df2['promo2_since_week'].astype(int).astype(str)
df2['promo_since'] = df2['promo_since'].apply(lambda x: datetime.datetime.strptime(x + '-1', '%Y-%W-%w') - datetime.timedelta(days=7))
df2['promo_time_week'] = ((df2['date'] - df2['promo_since']) / 7).apply(lambda x: x.days).astype(int)

df2 = df2.drop(['promo_since', 'competition_since'], axis=1)
```

- **State Holiday**:  
  - `'a'` → `'public_holiday'`

## 7. Análise Exploratória dos Dados

Qual é a importância da **Análise Exploratória de Dados (EDA)** em um projeto de Ciência de Dados?  

A Análise Exploratória é uma das etapas mais cruciais do processo, pois tem como principal objetivo **entender como as variáveis se relacionam com o fenômeno estudado**, neste caso, as vendas, e, **quantificar a força desse impacto**.  

Em essência, a EDA busca responder duas perguntas fundamentais:  
1. **Como as variáveis impactam o fenômeno?**  
2. **Qual a força desse impacto?**  

Além disso, ela permite **adquirir conhecimento sobre o negócio**, **validar hipóteses** e **gerar insights** que ajudam a compreender quais variáveis exercem maior influência sobre o resultado.  

A Análise Exploratória é geralmente dividida em **três etapas principais**:  
- **Análise Univariada:** foca em uma variável por vez, entendendo sua distribuição, comportamento e características individuais.  
- **Análise Bivariada:** estuda a relação entre duas variáveis, geralmente buscando correlações ou impactos diretos.  
- **Análise Multivariada:** analisa simultaneamente múltiplas variáveis para compreender interações mais complexas.  

Essa etapa não apenas guia as decisões de modelagem, como também **ajuda a formular hipóteses mais sólidas e estratégias mais eficazes** para o modelo preditivo.

```python
assortment_palette = {'basic': 'royalblue', 'extended': 'orange', 'extra': 'green'}

fig, axs = plt.subplots(3, 1, figsize=(15, 15))


assortment_sales_sum = df3[['assortment', 'sales']].groupby('assortment').sum().reindex(['basic', 'extended', 'extra'])
sns.barplot(
    x=assortment_sales_sum.index, 
    y=assortment_sales_sum['sales'], 
    ax=axs[0], 
    palette=[assortment_palette[a] for a in assortment_sales_sum.index]
)
axs[0].set_title('Barplot - Soma das Vendas por Assortment')
axs[0].set_xlabel('Assortment')
axs[0].set_ylabel('Vendas')
assortment_weeks = df3.groupby(['week_of_year', 'assortment'])['sales'].sum().reset_index()
for assortment in ['basic', 'extended', 'extra']:
    data = assortment_weeks[assortment_weeks['assortment'] == assortment]
    axs[1].plot(
        data['week_of_year'], 
        data['sales'], 
        label=assortment.capitalize(),
        color=assortment_palette[assortment]
    )

axs[1].set_title('Evolução das Vendas por Assortment ao Longo das Semanas')
axs[1].set_xlabel('Ano-Semana')
axs[1].set_ylabel('Vendas')
axs[1].legend(title='Assortment')

xticks_1 = axs[1].get_xticks()
xticklabels_1 = axs[1].get_xticklabels()
step_1 = max(1, len(xticklabels_1)//12)
axs[1].set_xticks(xticks_1[::step_1])
axs[1].set_xticklabels([label.get_text() for label in xticklabels_1][::step_1], rotation=90)

data_extra = assortment_weeks[assortment_weeks['assortment'] == 'extra']
axs[2].plot(
    data_extra['week_of_year'],
    data_extra['sales'],
    color=assortment_palette['extra'],
    label='Extra'
)
axs[2].set_title('Evolução das Vendas - Assortment Extra ao Longo das Semanas')
axs[2].set_xlabel('Ano-Semana')
axs[2].set_ylabel('Vendas')
axs[2].legend(title='Assortment')

xticks_2 = axs[2].get_xticks()
xticklabels_2 = axs[2].get_xticklabels()
step_2 = max(1, len(xticklabels_2)//12)
axs[2].set_xticks(xticks_2[::step_2])
axs[2].set_xticklabels([label.get_text() for label in xticklabels_2][::step_2], rotation=90)

plt.tight_layout()
plt.show()
```

**Análise:**  
A variável demonstra **dois comportamentos distintos**: enquanto lojas com sortimento básico ou estendido mantêm **vendas mais estáveis e elevadas**, aquelas com sortimento extra apresentam **queda significativa no volume de vendas**.

##### H2 Lojas com competidores mais próximos deveriam vender menos.

```python

aux1 = df3[['competition_distance', 'sales']].groupby('competition_distance').sum().reset_index()
aux2 = aux1[aux1['competition_distance'] != 200000.0].copy()


bin_edges = np.arange(0, 21000, 1000)
bin_edges = np.append(bin_edges, np.inf) 

labels = [f'{i}-{i+1000}' for i in range(0,20000,1000)] + ['>20000']

aux2['competition_distance_bin'] = pd.cut(
    aux2['competition_distance'],
    bins=bin_edges, 
    labels=labels,
    include_lowest=True
)

sales_by_distance = (
    aux2
    .groupby('competition_distance_bin', observed=False)['sales']
    .sum()
    .reindex(labels)
)


sales_by_distance_line = (
    aux2.groupby('competition_distance')['sales'].sum().sort_index()
)


num_bins = len(labels)
colors = plt.colormaps['viridis'](np.linspace(0, 1, num_bins))
colors = colors.tolist()


corr_matrix = aux1.corr(method='pearson')

fig = plt.figure(figsize=(20, 15))
grid = plt.GridSpec(2, 2, height_ratios=[2, 1], width_ratios=[3, 1], hspace=0.3, wspace=0.3)

ax0 = fig.add_subplot(grid[0, 0])
sns.barplot(
    x=sales_by_distance.index,
    y=sales_by_distance.values,
    ax=ax0,
    palette=colors,
    legend=False
)
ax0.set_title('Vendas por Distância da Concorrência')
ax0.set_xlabel('Distância da Concorrência')
ax0.set_ylabel('Vendas')
ax0.tick_params(axis='x', rotation=90)

axh = fig.add_subplot(grid[0, 1])
sns.heatmap(
    corr_matrix,
    annot=True, 
    fmt='.2f', 
    cmap='Purples', 
    vmin=-1, vmax=1, 
    square=True, 
    cbar=True, 
    ax=axh
)
axh.set_title('Heatmap da Correlação')

ax1 = fig.add_subplot(grid[1, :])
sns.kdeplot(
    data=aux2,
    x='competition_distance',
    weights='sales',
    fill=True,
    ax=ax1,
    color='orange'
)
ax1.set_title('Densidade das Vendas por Distância da Concorrência')
ax1.set_xlabel('Distância da Concorrência')
ax1.set_ylabel('Densidade das Vendas')
ax1.set_xlim(left=0, right=min(25000, aux2['competition_distance'].max()))
ax1.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

```

**Conclusão:** Falsa.  
As lojas com **competidores mais próximos** tendem, na verdade, a **vender mais**.  

**Análise:**  
O resultado indica que a **proximidade da concorrência** não necessariamente reduz as vendas, mas o contrário, pode estar associada a **regiões de maior movimento comercial**, onde há **maior concentração de clientes e demanda**.  
Essas áreas, mesmo com maior competição, tendem a gerar **maior fluxo de consumidores**, beneficiando todas as lojas ao redor.  

A variável apresenta um **valor estatístico mediano/baixo**, mas **comportamento consistente**, o que a torna **relevante para o modelo**, pois captura **efeitos indiretos de localização e densidade comercial**.

### 7.3 Analise Multivariada

Na Análise Multivariada, buscamos não apenas entender **como as variáveis explicam a resposta**, mas também **como elas se relacionam entre si**, com o objetivo de identificar **colinearidade forte**.  

Por que isso é importante? Modelos de machine learning seguem algumas premissas e boas práticas. Uma delas é conhecida como a **Navalha de Ockham**, que sugere que, entre modelos com desempenho similar, deve-se optar pelo **menos complexo**, pois modelos mais simples **tendem a generalizar melhor**.  

Uma forma de reduzir a complexidade do modelo é **diminuir o número de variáveis utilizadas no treinamento**, eliminando colunas que carregam informações redundantes. Para identificar essas redundâncias, analisamos **o grau de correlação entre as variáveis**, permitindo remover aquelas que são altamente correlacionadas e, assim, otimizar o desempenho do modelo.

#### 7.3.1 Variáveis Númericas

```python
numeric_columns = df3.select_dtypes(include=['int64', 'float64'])

corr = numeric_columns.corr()

plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Heatmap de Correlações entre Variáveis Numéricas')
plt.show()
```

#### 7.3.2 Categorical Attributes

```python
cat_cols = df3.select_dtypes(include=['object']).columns

cramers_results = pd.DataFrame(
    np.ones((len(cat_cols), len(cat_cols))),
    index=cat_cols,
    columns=cat_cols
)

for col1 in cat_cols:
    for col2 in cat_cols:
        if col1 == col2:
            cramers_results.loc[col1, col2] = 1.0
        else:
            confusion_matrix = pd.crosstab(df3[col1], df3[col2])
            if confusion_matrix.shape[0] > 1 and confusion_matrix.shape[1] > 1:
                cramers_results.loc[col1, col2] = cramers_v_bias_corrected(confusion_matrix)
            else:
                cramers_results.loc[col1, col2] = np.nan

plt.figure(figsize=(10,8))
sns.heatmap(cramers_results, annot=True, fmt='.2f', cmap='Blues', square=True, vmin=0, vmax=1)
plt.title("Heatmap de V de Cramer (correção de bias) entre variáveis categóricas")
plt.show()
```

### 8.1 Escala

Quando as variáveis possuem escalas diferentes, os modelos de machine learning tendem a **atribuir maior importância às variáveis com valores numéricos mais altos**, interpretando-as erroneamente como mais relevantes.  
Isso pode **enviesar o treinamento** e comprometer o desempenho do modelo.  

Por isso, é fundamental **padronizar as variáveis para uma mesma escala**, garantindo que **nenhuma delas se sobressaia apenas por causa da sua magnitude numérica**.  
Dessa forma, o modelo aprende com base nas relações reais entre as variáveis, e não nos seus valores absolutos.

#### 8.1.1 Normalização

A **normalização** (ou **padronização z-score**) é uma técnica de reescalonamento que transforma as variáveis para que tenham **média igual a 0** e **desvio padrão igual a 1**.  
Dessa forma, os dados são centralizados em torno da média e passam a ter uma escala comparável entre si.

Essa abordagem é especialmente útil para modelos que **assumem distribuição normal dos dados** ou que são **sensíveis à escala**, como *Regressão Linear*, *SVM* e *Redes Neurais*.

A fórmula utilizada é:

$$
x' = \frac{x - \mu}{\sigma}
$$

onde:  
- $x$ → valor original da variável  
- $\mu$ → média dos valores da variável  
- $\sigma$ → desvio padrão da variável  
- $x'$ → valor normalizado (padronizado)

Após a normalização, as variáveis passam a ter **distribuição centrada em 0** e **amplitude proporcional à sua variabilidade**, o que **melhora a estabilidade e a convergência do treinamento**.

No entanto, **no contexto deste projeto**, a normalização **não é viável**, pois **nenhuma das variáveis apresenta distribuição normal**.  
Aplicar esse tipo de transformação poderia **distorcer o comportamento original dos dados** e comprometer a interpretação dos resultados.

#### 8.1.2 Rescaling

##### Min-Max Scaler

O **Min-Max Scaler** é uma técnica de **reescalonamento** usada para ajustar os valores das variáveis para um intervalo específico — normalmente entre **0 e 1**. Ele **preserva a forma da distribuição original dos dados**, mas comprime todos os valores para caber nesse novo intervalo.

A fórmula do **Min-Max Scaler** é:

$$
x' = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
$$

onde:  
- $x$ é o valor original  
- $x_{\min}$ é o menor valor da variável  
- $x_{\max}$ é o maior valor da variável  
- $x'$ é o valor escalonado (após o reescalonamento)

O resultado é que:  
- O menor valor da variável passa a ser **0**  
- O maior valor passa a ser **1**  
- Todos os outros valores ficam proporcionalmente entre **0 e 1**

Essa técnica é útil quando:  
- Você quer **preservar a distribuição original** dos dados  
- O modelo é **sensível à escala** (como redes neurais, regressão logística e KNN)  
- Os dados **não seguem uma distribuição normal**, tornando a normalização (z-score) menos apropriada  

> **Atenção:** O **Min-Max Scaler é sensível a outliers.** Valores muito altos ou muito baixos podem distorcer a escala e influenciar o reescalonamento das demais observações.

```python
# Reescalando promo_time_week
mm_scaler_promo_time_week = MinMaxScaler()
df4['promo_time_week'] = mm_scaler_promo_time_week.fit_transform(df4[['promo_time_week']])
with open('../api/rossman/parameter/minmax_scaler_promo_time_week.pkl', 'wb') as f:
    pickle.dump(mm_scaler_promo_time_week, f)

# Reescalando year
mm_scaler_year = MinMaxScaler()
df4['year'] = mm_scaler_year.fit_transform(df4[['year']])
with open('../api/rossman/parameter/minmax_scaler_year.pkl', 'wb') as f:
    pickle.dump(mm_scaler_year, f)

```

##### Robust Scaler

O **Robust Scaler** é uma técnica de **reescalonamento** usada para ajustar os valores das variáveis com base em estatísticas **robustas** (mediana e intervalo interquartílico), tornando-o **menos sensível a outliers**. Ele **centraliza os dados na mediana** e os escala de acordo com o **IQR (intervalo entre o 1º e 3º quartil)**.

A fórmula do **Robust Scaler** é:

$$
x' = \frac{x - \text{mediana}(x)}{\text{IQR}(x)}
$$

onde:  
- $x$ é o valor original  
- $\text{mediana}(x)$ é a mediana da variável  
- $\text{IQR}(x)$ é o intervalo interquartílico (Q3 - Q1)  
- $x'$ é o valor escalonado (após o reescalonamento)

O resultado é que:  
- A mediana da variável passa a ser **0**  
- Valores próximos à mediana ficam entre **-1 e 1** (aproximadamente)  
- Outliers **não distorcem** tanto a escala dos demais valores

Essa técnica é útil quando:  
- Há **outliers significativos** nos dados  
- Você quer **preservar a distribuição relativa** dos dados sem ser influenciado por valores extremos  
- O modelo é **sensível à escala** (como regressão linear, redes neurais e KNN), mas você não quer que outliers distorçam o treino

> **Atenção:** O **Robust Scaler não limita os valores a um intervalo fixo** como o Min-Max. Valores extremos podem ainda gerar números maiores ou menores que 1, mas a influência deles na escala geral é reduzida.


```python
scaler = RobustScaler()
# Reescalando competition_distance
df4['competition_distance'] = scaler.fit_transform(df4[['competition_distance']])
with open('../api/rossman/parameter/robust_scaler_competition_distance.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# Reescalando competition_time_month
scaler2 = RobustScaler()
df4['competition_time_month'] = scaler2.fit_transform(df4[['competition_time_month']])
with open('../api/rossman/parameter/robust_scaler_competition_time_month.pkl', 'wb') as f:
    pickle.dump(scaler2, f)
```

##### Pickle

Você deve ter notado no código o salvamento de arquivos no formato **pickle**. Mas por que isso é importante?  

Ao aplicar transformações nos dados, como **reescalonamento** (Min-Max, Robust, etc.), o modelo aprende a fazer previsões considerando essa escala específica. Para que futuras previsões sejam consistentes, é fundamental **salvar o objeto que realizou a transformação** (por exemplo, o scaler) e aplicá-lo novamente aos novos dados antes de usar o modelo.  

O **pickle** permite armazenar tanto o modelo quanto os objetos de pré-processamento, garantindo que:  
- A escala dos dados seja **reproduzida corretamente**  
- O modelo possa ser **reutilizado** sem precisar treinar novamente  
- Novas previsões sejam **coerentes** com o treinamento original

#### 8.3.2 Transformação da Variável Resposta

Como já vimos, a variável **sales** não possui uma distribuição normal. Para que o modelo funcione melhor, é recomendado aplicar uma **transformação de grandeza**, tornando a distribuição da variável o mais próxima possível de uma **distribuição normal**.

Uma das transformações mais comuns é a **transformação logarítmica**.

- Ela consiste em aplicar o logaritmo aos valores da variável:  

$$
y' = \log(y + 1)
$$

onde:

- O **+1** é usado para evitar problemas com valores zero (pois \(\log(0)\) é indefinido)

**Vantagens:**  
- Reduz a **assimetria** de distribuições muito concentradas em valores baixos e com cauda longa.  
- Comprime grandes valores, evitando que eles dominem o aprendizado do modelo.  
- Facilita que modelos assumam **resíduos aproximadamente normais**, melhorando previsões.

**Observação:**  
- Após a predição, se for necessário interpretar os valores no **espaço original**, deve-se aplicar a **transformação inversa**:

$$
y = \exp(y') - 1
$$

#### 8.3.3 Transformação de Natureza

Algumas variáveis possuem uma **natureza cíclica**, ou seja, elas **se repetem em intervalos regulares**. Exemplos comuns incluem:  
- Dias da semana (1 = domingo, 7 = sábado)  
- Meses do ano (1 = janeiro, 12 = dezembro)  
- Horas do dia (0–23)  

Se tratarmos essas variáveis como números comuns, o modelo pode interpretar erroneamente a distância entre os valores. Por exemplo, no caso dos meses, o modelo poderia pensar que dezembro (12) está longe de janeiro (1), quando na realidade eles são consecutivos.

**Como aplicar a transformação cíclica?**

Para preservar a natureza circular da variável, podemos usar funções trigonométricas **seno** e **cosseno** para criar duas novas colunas:

$$
x_{\text{sin}} = \sin\left(\frac{2 \pi x}{\max(x)}\right)
$$

$$
x_{\text{cos}} = \cos\left(\frac{2 \pi x}{\max(x)}\right)
$$


onde:  
- x é o valor original da variável cíclica  
- max(x) é o valor máximo da variável (por exemplo, 7 para dias da semana, 12 para meses)  
- x_sin e x_cos são os valores transformados que preservam a relação cíclica

**Vantagens:**  
- Mantém a **proximidade correta** entre valores no final e no início do ciclo  
- Facilita que o modelo capture **padrões sazonais ou periódicos**  
- Evita a interpretação errada de distância entre valores cíclicos


```python
# Transformação de seno e cosseno para a variável cíclica 'month'
df4['month_sin'] = np.sin(2 * np.pi * df4['month'] / 12)
df4['month_cos'] = np.cos(2 * np.pi * df4['month'] / 12)
df5 = df4.drop('month', axis=1)

# Transformação de seno e cosseno para a variável cíclica 'day'
df4['day_sin'] = np.sin(2 * np.pi * df5['day'] / 31)
df4['day_cos'] = np.cos(2 * np.pi * df5['day'] / 31)
df4 = df4.drop('day', axis=1)

# Transformação de seno e cosseno para a variável cíclica 'day_of_week'
df4['day_of_week_sin'] = np.sin(2 * np.pi * df4['day_of_week'] / 7)
df4['day_of_week_cos'] = np.cos(2 * np.pi * df4['day_of_week'] / 7)
df4 = df4.drop('day_of_week', axis=1)

# Transformação de seno e cosseno para a variável cíclica 'week_of_year'
df4['week_of_year_sin'] = np.sin(2 * np.pi * df4['week_of_year'] / 52)
df4['week_of_year_cos'] = np.cos(2 * np.pi * df4['week_of_year'] / 52)
df4 = df4.drop('week_of_year', axis=1)

```

### 9.1 Boruta como seletor de variáveis

O **Boruta** é um algoritmo de seleção de variáveis baseado em **modelos de árvores de decisão**, geralmente utilizando o **Random Forest** como estimador principal.  
Seu objetivo é identificar **todas as variáveis relevantes** para a predição, e não apenas um subconjunto mínimo.

**Como o Boruta funciona?**

1. **Criação de variáveis sombreadas (*shadow features*)**  
   - O algoritmo começa duplicando o conjunto de variáveis originais.  
   - Em seguida, embaralha os valores das cópias (as *shadow features*), removendo qualquer relação com a variável alvo.  
   - Isso cria um **padrão de referência**: variáveis totalmente aleatórias.

2. **Treinamento do modelo base (Random Forest)**  
   - O Boruta treina um modelo de Random Forest com o conjunto expandido (variáveis reais + sombreadas).  
   - O modelo atribui uma **importância** para cada variável.

3. **Comparação entre variáveis reais e sombreadas**  
   - O algoritmo compara a importância das variáveis reais com a **melhor variável sombreada**.  
   - Se uma variável real tiver importância significativamente maior, ela é **confirmada** como relevante.  
   - Se for menor, é **rejeitada**.  
   - As variáveis indecisas passam por **novas iterações**.

4. **Iteração até convergência**  
   - O processo se repete várias vezes até que todas as variáveis sejam classificadas como **aceitas** ou **rejeitadas**, ou até que o número máximo de iterações seja atingido.

### 10.1 Dividindo o dataframe em treino e teste

**Por que dividir o DataFrame?**

Ao treinar um modelo com certos dados e realizar previsões utilizando **esses mesmos dados**, é natural que o modelo apresente uma **ótima performance**, afinal, ele já “viu” esses exemplos durante o treinamento.  

Esse comportamento, porém, **não reflete a realidade**. Em um ambiente de produção, o modelo será exposto a dados **novos**, que ele nunca encontrou antes.  

Por isso, é fundamental dividir o conjunto de dados em duas partes principais:

- **Treino (train):** usado para ajustar os parâmetros do modelo e ensinar os padrões existentes nos dados.  
- **Teste (test):** usado para avaliar o desempenho do modelo em dados desconhecidos, simulando sua aplicação real.

Essa prática garante que possamos medir **a capacidade de generalização** do modelo, evitando o fenômeno de **overfitting**, quando ele se torna bom demais apenas nos dados de treino, mas falha em novos casos.

```python
df6_sorted = df6.sort_values('date')

# Descobrir a última data disponível
last_date = df6_sorted['date'].max()
first_test_date = last_date - pd.Timedelta(weeks=6)

# Definir conjunto de teste e treino
test_mask = df6_sorted['date'] > first_test_date

X = df6_sorted.drop(['sales'], axis=1)
y = df6_sorted['sales']

X_train = X[~test_mask].drop('date', axis=1)
y_train = y[~test_mask]
X_test = X[test_mask].drop('date', axis=1)
y_test = y[test_mask]
```

### 10.2 Métricas de Avaliação

Para comparar modelos de **Machine Learning**, utilizamos métricas que indicam **quão bem o modelo está prevendo os valores da variável resposta**. As mais comuns para problemas de regressão (como previsão de vendas) são:

---

#### 1. MAE (Mean Absolute Error)  

O **MAE** mede o erro médio absoluto entre os valores reais e as previsões:

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$


- $y_i$ = valor real  
- $\hat{y}_i$ = valor previsto pelo modelo  
- $n$ = número de observações

> Interpretação: o MAE indica, em média, **quanto o modelo erra**, em unidades da variável resposta.  

---

#### 2. RMSE (Root Mean Squared Error)  

O **RMSE** mede o erro quadrático médio e dá **mais peso a grandes erros**:

$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

- Assim como o MAE, quanto menor o RMSE, melhor.  
- É sensível a **outliers**, já que erros maiores são amplificados pelo quadrado.

---

#### 3. MAPE (Mean Absolute Percentage Error)  

O **MAPE** calcula o erro absoluto médio em **percentual**, facilitando a interpretação relativa:

$$
MAPE = \frac{100}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|
$$

- Útil para medir desempenho em unidades **relativas**, especialmente quando valores variam em escalas diferentes.  
- Limitação: não funciona bem quando \(y_i = 0\), pois gera divisão por zero.

---

#### 4. R² (Coeficiente de Determinação)  

O **R²** indica a proporção da **variância explicada pelo modelo** em relação à variância total da variável resposta:

$$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$

- $\bar{y} = \text{média dos valores reais}$
- Varia de **0 a 1** (quanto mais próximo de 1, melhor o modelo explica os dados)  
- Valores negativos podem ocorrer se o modelo for **pior que simplesmente prever a média**.

---

Essas métricas permitem **comparar modelos diferentes** e identificar aquele que melhor representa os dados, considerando tanto erros absolutos, relativos e a capacidade de explicar a variabilidade dos valores reais.

### 10.3 Average Model

Um **Average Model** é o tipo mais simples de modelo possível, algo que qualquer pessoa poderia calcular, mesmo sem o uso de um computador.  

Ele consiste basicamente em **utilizar a média dos valores observados** (neste caso, as vendas) como uma previsão constante.  

Em outras palavras, o modelo assume que todas as futuras observações terão um valor igual à média histórica das vendas.

Apesar de sua simplicidade, o Average Model desempenha um papel importante no processo de modelagem:  
ele serve como uma **linha de base (baseline)** para comparar o desempenho de modelos mais complexos.  

Se um modelo de *Machine Learning* não for capaz de superar a performance de um simples Average Model, então ele **não está realmente aprendendo padrões relevantes** nos dados.

### 10.3 Linear Regression Model


O modelo de **Regressão Linear** é um dos métodos mais simples e interpretáveis de aprendizado supervisionado. Seu objetivo é encontrar uma relação linear entre as variáveis independentes (features) e a variável dependente (resposta).

Em termos matemáticos, o modelo assume a seguinte forma:

$$
y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \varepsilon
$$

Onde:  
- $y$ → variável alvo (no caso, as vendas da loja)  
- $x₁, x₂, …, xₙ$ → variáveis explicativas (ex: promoções, feriados, concorrência, etc.)  
- $β₀$ → intercepto (valor de *y* quando todas as variáveis são zero)  
- $β₁, β₂, …, βₙ$ → coeficientes do modelo, que indicam quanto a variável resposta muda quando a variável correspondente aumenta em uma unidade, mantendo as demais constantes  
- $ε$ → erro aleatório, representando a variação não explicada pelo modelo  

O treinamento do modelo consiste em encontrar os valores de $β$ que minimizam o erro entre as predições e os valores reais, geralmente medido pelo **erro quadrático médio (MSE)**.

Apesar de sua simplicidade, a Regressão Linear é poderosa para compreender **relações lineares** entre variáveis e serve como uma boa linha de base antes de testar modelos mais complexos. Ela também permite interpretar de forma clara a **importância e direção dos efeitos** das variáveis, ou seja, se cada variável aumenta ou reduz as vendas e em qual magnitude.

No entanto, o modelo apresenta limitações quando as relações entre as variáveis não são lineares ou quando existe **multicolinearidade** (forte correlação entre as variáveis independentes), o que pode distorcer os coeficientes e reduzir a performance preditiva.


```python

# Treina o modelo de Regressão Linear
lr = LinearRegression()
lr.fit(X_train, y_train)

# Faz previsões no conjunto de teste
y_pred_lr = lr.predict(X_test)

# Rescale para o domínio original
y_test_rescaled = np.expm1(y_test_array)
y_pred_lr_rescaled = np.expm1(y_pred_lr)

# Calcula as métricas de avaliação
df_metrics_lr = calcular_metricas(y_test_rescaled, y_pred_lr_rescaled, 'LinearRegression')
df_metrics_lr
```

### 10.4 Linear Regression Regularized Model - Lasso

A **Regressão Linear Regularizada** é uma variação da regressão linear tradicional que busca evitar o **overfitting**, ou seja, quando o modelo se ajusta demais aos dados de treino e perde capacidade de generalização. Para isso, adiciona-se um **termo de penalização** na função de custo, que controla o tamanho dos coeficientes do modelo.

O **Lasso (Least Absolute Shrinkage and Selection Operator)** utiliza a **penalização L1**, que adiciona à função de custo a soma dos valores absolutos dos coeficientes. Sua função de custo é dada por:

$$
J(\beta) = \frac{1}{2n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2 + \alpha \sum_{j=1}^{p} |\beta_j|
$$

Onde:  
- $n$ → número de observações  
- $p$ → número de variáveis  
- $α$ → parâmetro de regularização (controla o peso da penalização)  
- $ βⱼ $ → coeficiente da variável $j$  

A penalização L1 tem um efeito importante: **força alguns coeficientes a se tornarem exatamente zero**, o que faz com que o Lasso realize uma **seleção automática de variáveis**.  
Ou seja, além de reduzir o overfitting, ele também ajuda a **identificar quais variáveis são mais relevantes** para o modelo.


```python
# Instancia e treina o Lasso
lasso = Lasso(alpha=0.01, random_state=42)
lasso.fit(X_train, y_train)

# Faz previsões
y_pred_lasso = lasso.predict(X_test)

# Rescale para o domínio original
y_test_rescaled = np.expm1(y_test_array)
y_pred_lasso_rescaled = np.expm1(y_pred_lasso)

# Calcula as métricas
df_metrics_lasso = calcular_metricas(y_test_rescaled, y_pred_lasso_rescaled, 'Lasso')
df_metrics_lasso
```

### 10.5 Linear Regression Regularized Model - Ridge

O **Ridge Regression** utiliza a **penalização L2**, que adiciona o **quadrado dos coeficientes** à função de custo:

$$
J(\beta) = \frac{1}{2n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2 + \alpha \sum_{j=1}^{p} \beta_j^2
$$

Diferente do Lasso, o Ridge **não zera os coeficientes**, mas os **reduz de forma proporcional**. Isso é útil quando há **multicolinearidade**, isto é, quando variáveis independentes estão fortemente correlacionadas entre si.  
O Ridge distribui melhor o peso entre as variáveis correlacionadas, reduzindo a variância do modelo e aumentando sua estabilidade.


```python
# Instancia e treina o Ridge
ridge = Ridge(alpha=0.01, random_state=42)
ridge.fit(X_train, y_train)

# Faz previsões
y_pred_ridge = ridge.predict(X_test)

# Rescale para o domínio original
y_test_rescaled = np.expm1(y_test_array)
y_pred_ridge_rescaled = np.expm1(y_pred_ridge)

# Calcula as métricas
df_metrics_ridge = calcular_metricas(y_test_rescaled, y_pred_ridge_rescaled, 'Ridge')
df_metrics_ridge
```

### 10.6 Random Forest Regressor

O **Random Forest Regressor** é um modelo de aprendizado de máquina baseado em **múltiplas árvores de decisão**. Ele pertence à classe dos modelos de **ensemble**, ou seja, modelos que combinam vários estimadores simples para produzir um resultado mais robusto e estável.

A ideia central é simples:  
em vez de depender de uma única árvore de decisão, que pode ser instável e sensível a pequenas variações nos dados, o Random Forest constrói **diversas árvores independentes** e, em seguida, **combina suas previsões**.  
Para regressão, o resultado final é geralmente a **média das previsões** feitas por todas as árvores.

---

**Como o modelo funciona?**

1. **Bootstrap Sampling (Amostragem com reposição)**  
   O conjunto de dados de treino é dividido aleatoriamente em vários subconjuntos (chamados *bootstraps*), e cada árvore é treinada em um desses subconjuntos.

2. **Seleção aleatória de variáveis (Feature Subsampling)**  
   Em cada divisão de uma árvore, apenas um subconjunto aleatório das variáveis é considerado. Isso garante **diversidade entre as árvores**, evitando que todas aprendam os mesmos padrões.

3. **Treinamento das árvores**  
   Cada árvore tenta prever o valor da variável resposta (no caso, `sales`) com base nos dados que recebeu.

4. **Agregação dos resultados**  
   Para regressão, o Random Forest calcula a **média das previsões** de todas as árvores.  
   Assim, reduz o **erro de variância** e torna o modelo mais **generalizável**.


**Quais as vantagens?**

- Reduz o risco de **overfitting** em comparação a uma árvore única.  
- Trabalha bem com **dados não lineares**.  
- É robusto a **outliers e ruído nos dados**.  
- Fornece medidas de **importância das variáveis (feature importance)**, permitindo interpretar quais atributos mais influenciam as predições.

**Quais as desvantagens?**

- O modelo é mais **lento** para treinar e prever quando comparado a uma única árvore.  
- É menos interpretável. Não é fácil visualizar ou entender todas as decisões internas.  
- Pode precisar de ajuste fino de hiperparâmetros (como número de árvores, profundidade máxima, número de variáveis por divisão, etc.).

**Fórmula conceitual**

A predição final de uma Random Forest para uma entrada \( x \) é a média das predições das \( N \) árvores individuais:

$$
\hat{y} = \frac{1}{N} \sum_{i=1}^{N} T_i(x)
$$

Onde:  
- $ \hat{y} $ → valor previsto pelo modelo  
- $ N $ → número total de árvores  
- $ T_i(x) $ → predição da *i*-ésima árvore para a entrada $ x $


Em resumo, o **Random Forest Regressor** é um dos modelos mais poderosos e versáteis para problemas de regressão, oferecendo excelente desempenho mesmo sem um grande pré-processamento ou ajustes complexos.

```python
# Instancia e treina o Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=1)
rf.fit(X_train, y_train)

# Faz previsões
y_pred_rf = rf.predict(X_test)

# Rescale para o domínio original
y_test_rescaled = np.expm1(y_test_array)
y_pred_rf_rescaled = np.expm1(y_pred_rf)

# Calcula as métricas
df_metrics_rf = calcular_metricas(y_test_rescaled, y_pred_rf_rescaled, 'RandomForest')
df_metrics_rf
```

### 10.7 XGBoost Regressor

O **XGBoost (Extreme Gradient Boosting)** é um dos algoritmos de **ensemble** mais poderosos e populares para tarefas de **regressão e classificação**.  
Ele é baseado no conceito de **Gradient Boosting**, uma técnica que combina o resultado de vários modelos fracos (normalmente árvores de decisão simples) para formar um modelo forte e preciso.



**Intuição do modelo**

Enquanto o **Random Forest** cria várias árvores **independentes** e depois combina suas previsões, o **XGBoost** constrói as árvores **sequencialmente**.  
Cada nova árvore tenta **corrigir os erros** cometidos pelas árvores anteriores.

O processo pode ser entendido como um aprendizado passo a passo, onde o modelo melhora a cada iteração.



**Como o modelo funciona?**

1. O XGBoost começa com uma predição inicial (geralmente a média dos valores da variável resposta).  
2. Ele calcula o **erro (resíduo)** entre as predições e os valores reais.  
3. Cria uma nova árvore para tentar prever esses resíduos.  
4. A nova árvore é adicionada ao modelo, ajustando as predições anteriores.  
5. O processo se repete até atingir o número máximo de árvores ou até que os erros parem de diminuir.



**Fórmula conceitual**

A predição final de um modelo XGBoost para uma entrada \( x \) é dada por:

$$
\hat{y}_i = \sum_{k=1}^{K} f_k(x_i)
$$

Onde:  
- $ \hat{y}_i $ → valor previsto para a observação $ i $ 
- $ K $ → número total de árvores  
- $ f_k $ → função da $k$-ésima árvore  

Cada nova árvore é treinada para minimizar uma **função de perda** usando o gradiente dos erros (daí o nome *Gradient Boosting*).


**Função de Custo**

O XGBoost minimiza uma função de custo que combina o erro de predição com uma penalização pela complexidade das árvores:

$$
Obj = \sum_{i=1}^{n} l(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)
$$

Onde:  
- $ l(y_i, \hat{y}_i) $ → mede o erro entre o valor real e o previsto  
- $ \Omega(f_k) $ → penaliza árvores muito complexas, evitando **overfitting**



**Quais as vantagens?**

- Extremamente **rápido e eficiente**, com suporte a **paralelização**.  
- Possui **regularização embutida (L1 e L2)**, reduzindo o overfitting.  
- Alta performance em problemas com **grandes volumes de dados**.  
- Permite análise de **importância das variáveis** (*feature importance*).  
- Funciona bem com dados **não lineares** e com **interações complexas**.



**Quais as desvantagens?**

- Exige **mais ajuste de hiperparâmetros** (tuning) para alcançar o melhor desempenho.  
- Pode ser **mais difícil de interpretar** que modelos lineares.  
- Se mal configurado, pode **superajustar (overfit)** aos dados de treino.


O **XGBoost Regressor** é uma evolução do método de *Gradient Boosting*, projetado para ser **rápido, regularizado e altamente escalável**.  
Por isso, é amplamente utilizado em competições de ciência de dados e em sistemas de previsão no mundo real.

### 10.9 Cross Validation

Para obter uma avaliação mais confiável, é utilizada a técnica de **Cross Validation (Validação Cruzada)**.  
Ela consiste em dividir o conjunto de dados em múltiplas partes (ou *folds*), e o modelo é treinado e testado várias vezes, alternando qual parte é usada como teste.

O método mais comum é o **K-Fold Cross Validation**, onde o dataset é dividido em $ K $ partes (geralmente 5 ou 10).  
O modelo é treinado $ K $ vezes, e a métrica final é a **média das performances** de cada iteração.

Para este conjunto de dados, que depende diretamente da **sequência temporal das observações**, não é apropriado utilizar a validação cruzada tradicional (*K-Fold*), pois ela **embaralha os dados** e pode fazer com que o modelo "veja o futuro" durante o treino.  
Isso violaria o princípio de causalidade em séries temporais, resultando em uma avaliação irreal do desempenho do modelo.

Em problemas de previsão de vendas (como o caso da Rossmann), o objetivo é **prever o futuro com base no passado**.  
Portanto, o modelo deve ser treinado em períodos anteriores e testado em períodos posteriores, respeitando a ordem cronológica dos dados.

A **validação por janela de tempo** resolve esse problema criando divisões sucessivas onde cada janela de treino é seguida por uma janela de validação mais recente.

## 11. Fine Tuning

Após identificar o modelo com melhor desempenho, neste caso, o **XGBoost Regressor**. O próximo passo é realizar o **Fine Tuning**, ou **ajuste fino dos hiperparâmetros**.  
Essa etapa é fundamental para **otimizar o desempenho** do modelo e garantir que ele atinja seu **máximo potencial preditivo**.

O **Fine Tuning** consiste em ajustar cuidadosamente os **hiperparâmetros** do modelo.  
Os hiperparâmetros são configurações externas ao processo de treinamento, ou seja, **valores definidos antes do modelo aprender com os dados**.  
Diferentemente dos parâmetros internos (como coeficientes de uma regressão), os hiperparâmetros **não são aprendidos automaticamente**, eles precisam ser configurados pelo analista ou encontrados por meio de busca sistemática.

O objetivo do Fine Tuning é encontrar a **melhor combinação de hiperparâmetros** para maximizar o desempenho do modelo nas métricas escolhidas (ex: RMSE, MAE, MAPE ou R²).

### 11.1 Utilizando a busca bayesiana

A **busca bayesiana** é uma técnica de otimização de hiperparâmetros que utiliza princípios de **probabilidade e estatística** para encontrar, de forma mais eficiente, os melhores valores de configuração de um modelo de machine learning.

Diferente da **busca em grade (Grid Search)**, que testa todas as combinações possíveis de hiperparâmetros, ou da **busca aleatória (Random Search)**, que testa combinações escolhidas aleatoriamente, a busca bayesiana **aprende com os resultados anteriores** e **estima quais combinações têm maior probabilidade de melhorar o desempenho** do modelo.

**Funcionamento básico**

1. Define-se um **espaço de busca** dos hiperparâmetros (intervalos de valores possíveis).  
2. A busca começa testando algumas combinações iniciais aleatórias.  
3. Com base nos resultados obtidos (por exemplo, RMSE ou R²), o algoritmo cria um **modelo probabilístico** que representa a relação entre os hiperparâmetros e o desempenho do modelo.  
4. Esse modelo é usado para **escolher inteligentemente** as próximas combinações a serem testadas, priorizando aquelas com **maior chance de melhorar** o resultado.  
5. O processo é repetido até atingir o número máximo de iterações ou até a melhora no desempenho estabilizar.

```python

# Espaço de busca para busca bayesiana
param_search = {
    'n_estimators': Integer(500, 3000),
    'max_depth': Integer(3, 15),
    'learning_rate': Real(0.005, 0.105, prior='uniform'),
    'subsample': Real(0.6, 1.0, prior='uniform'),
    'colsample_bytree': Real(0.6, 1.0, prior='uniform'),
    'gamma': Real(0.0, 0.3, prior='uniform'),
    'min_child_weight': Integer(1, 10)
}

xgb_reg = XGBRegressor(objective='reg:squarederror', random_state=42)


bayes_search = BayesSearchCV(
    estimator=xgb_reg,
    search_spaces=param_search,
    n_iter=30,
    scoring='r2',
    n_jobs=1,
    cv=3,
    verbose=2,
    random_state=42
)


X = df7.drop(['sales', 'date'], axis=1)
y = df7['sales']

bayes_search.fit(X, y)

print("Melhores parâmetros encontrados:", bayes_search.best_params_)
print("Melhor R²:", bayes_search.best_score_)
```

### 13.2 Pipeline de dados

```python
import pickle
import pandas as pd
import numpy as np
import datetime
import os


class Rossman:
    """
    Classe principal para gerenciamento do modelo Rossman, incluindo limpeza de dados,
    engenharia de atributos, preparação dos dados e predição.

    Esta classe encapsula todo o pipeline de preparação e inferência, utilizando os mesmos 
    scalers e encoders salvos durante o treinamento para garantir consistência no deploy.
    """

    expected_features = [
        'store',
        'promo',
        'competition_distance',
        'promo2',
        'promo2_since_week',
        'competition_open_since_month',
        'competition_time_month',
        'promo_time_week',
        'month_cos',
        'month_sin',
        'day_sin',
        'day_cos',
        'day_of_week_sin',
        'day_of_week_cos',
        'week_of_year_sin',
        'week_of_year_cos',
        'store_type_encoded',
        'assortment_encoded'
    ]

    def __init__(self):
        base_dir = os.path.dirname(os.path.abspath(__file__))  # pasta api/rossman

        with open(os.path.join(base_dir, 'parameter', 'robust_scaler_competition_distance.pkl'), 'rb') as f:
            self.scaler_competition_distance = pickle.load(f)

        with open(os.path.join(base_dir, 'parameter', 'robust_scaler_competition_time_month.pkl'), 'rb') as f:
            self.scaler_competition_time_month = pickle.load(f)

        with open(os.path.join(base_dir, 'parameter', 'minmax_scaler_promo_time_week.pkl'), 'rb') as f:
            self.scaler_promo_time_week = pickle.load(f)

        with open(os.path.join(base_dir, 'parameter', 'minmax_scaler_year.pkl'), 'rb') as f:
            self.scaler_year = pickle.load(f)

        with open(os.path.join(base_dir, 'parameter', 'ohe_state_holiday.pkl'), 'rb') as f:
            self.ohe_state_holiday = pickle.load(f)

        with open(os.path.join(base_dir, 'parameter', 'le_store_type.pkl'), 'rb') as f:
            self.le_store_type = pickle.load(f)

        with open(os.path.join(base_dir, 'parameter', 'assortment_mapping.pkl'), 'rb') as f:
            self.assortment_mapping = pickle.load(f)
        
        model_path = os.path.join(base_dir, 'model', 'model_rossman_sales.pkl')
        model_path = os.path.abspath(model_path)
        
        with open(model_path, 'rb') as f:
            self.model = pickle.load(f)



    def clean_data(self, df):
        df = df.copy()
        def snake_case(lst):
            def convert(s):
                s = s.replace(' ', '_')
                new_s = ""
                for i, c in enumerate(s):
                    if c.isupper():
                        if i > 0 and (s[i-1].islower() or (i+1 < len(s) and s[i+1].islower())):
                            new_s += "_"
                        new_s += c.lower()
                    else:
                        new_s += c
                return new_s
            return [convert(s) for s in lst]
        df.columns = snake_case(df.columns)
        df['date'] = pd.to_datetime(df['date'])
        df['competition_distance'] = df['competition_distance'].fillna(200000)
        df['competition_open_since_month'] = df['competition_open_since_month'].fillna(df['date'].dt.month)
        df['competition_open_since_year'] = df['competition_open_since_year'].fillna(df['date'].dt.year)
        df['promo2_since_week'] = df['promo2_since_week'].fillna(df['date'].dt.isocalendar().week)
        df['promo2_since_year'] = df['promo2_since_year'].fillna(df['date'].dt.year)
        df['promo_interval'] = df['promo_interval'].fillna(0)
        df['competition_open_since_month'] = df['competition_open_since_month'].astype(int)
        df['competition_open_since_year'] = df['competition_open_since_year'].astype(int)
        df['promo2_since_week'] = df['promo2_since_week'].astype(int)
        df['promo2_since_year'] = df['promo2_since_year'].astype(int)
        return df

    def feature_engineering(self, df):
        df = df.copy()
        month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sept', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
        df['month_map'] = df['date'].dt.month.map(month_map)
        df['is_promo'] = df.apply(lambda row: 1 if str(row['month_map']) in str(row['promo_interval']) else 0, axis=1)
        df.drop('month_map', axis=1, inplace=True)
        df['year'] = df['date'].dt.year.astype(int)
        df['month'] = df['date'].dt.month.astype(int)
        df['day'] = df['date'].dt.day.astype(int)
        df['week_of_year'] = df['date'].dt.isocalendar().week.astype(int)
        df['competition_since'] = pd.to_datetime(
            dict(year=df['competition_open_since_year'],
                 month=df['competition_open_since_month'],
                 day=1)
        )
        df['competition_time_month'] = ((df['date'] - df['competition_since']).dt.days / 30).round().astype(int)
        df['promo_since'] = df['promo2_since_year'].astype(str) + '-' + df['promo2_since_week'].astype(str)
        df['promo_since'] = df['promo_since'].apply(
            lambda x: datetime.datetime.strptime(x + '-1', '%Y-%W-%w') - datetime.timedelta(days=7)
        )
        df['promo_time_week'] = ((df['date'] - df['promo_since']) / 7).apply(lambda x: x.days).astype(int)
        if 'assortment' in df.columns:
            df['assortment'] = df['assortment'].map({'a': 'basic', 'b': 'extra', 'c': 'extended'})
        if 'state_holiday' in df.columns:
            df['state_holiday'] = df['state_holiday'].map(
                {'a': 'public holiday', 'b': 'easter holiday', 'c': 'christmas', '0': 'regular_day'}
            )
        if 'month_cos' not in df.columns:
            df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
        if 'month_sin' not in df.columns:
            df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
        if 'day_cos' not in df.columns:
            df['day_cos'] = np.cos(2 * np.pi * df['day'] / 31)
        if 'day_sin' not in df.columns:
            df['day_sin'] = np.sin(2 * np.pi * df['day'] / 31)
        if 'day_of_week' in df.columns:
            if 'day_of_week_cos' not in df.columns:
                df['day_of_week_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
            if 'day_of_week_sin' not in df.columns:
                df['day_of_week_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
        if 'week_of_year_cos' not in df.columns:
            df['week_of_year_cos'] = np.cos(2 * np.pi * df['week_of_year'] / 52)
        if 'week_of_year_sin' not in df.columns:
            df['week_of_year_sin'] = np.sin(2 * np.pi * df['week_of_year'] / 52)
        drop_cols = [
            'customers',
            'open',
            'promo_interval',
            'promo2_since_year',
            'competition_since',
            'promo_since'
        ]
        drop_cols = [col for col in drop_cols if col in df.columns]
        if drop_cols:
            df = df.drop(drop_cols, axis=1)
        return df

    def data_preparation(self, df):
        df = df.copy()
        if 'competition_distance' in df.columns:
            df['competition_distance'] = self.scaler_competition_distance.transform(df[['competition_distance']])
        if 'competition_time_month' in df.columns:
            df['competition_time_month'] = self.scaler_competition_time_month.transform(df[['competition_time_month']])
        if 'promo_time_week' in df.columns:
            df['promo_time_week'] = self.scaler_promo_time_week.transform(df[['promo_time_week']])
        if 'year' in df.columns:
            df['year'] = self.scaler_year.transform(df[['year']])
        if 'state_holiday' in df.columns:
            state_holiday_encoded = self.ohe_state_holiday.transform(df[['state_holiday']])
            state_holiday_columns = self.ohe_state_holiday.get_feature_names_out(['state_holiday'])
            df_state_holiday = pd.DataFrame(state_holiday_encoded, columns=state_holiday_columns, index=df.index)
            df = pd.concat([df.drop('state_holiday', axis=1), df_state_holiday], axis=1)
        if 'store_type' in df.columns:
            df['store_type_encoded'] = self.le_store_type.transform(df['store_type'])
        if 'assortment' in df.columns:
            df['assortment_encoded'] = df['assortment'].map(self.assortment_mapping)
        drop_cats = [c for c in ['assortment', 'store_type'] if c in df.columns]
        if drop_cats:
            df = df.drop(drop_cats, axis=1)
        missing_cols = [col for col in self.expected_features if col not in df.columns]
        if missing_cols:
            raise ValueError(f"Missing expected features: {missing_cols}")
        df = df[self.expected_features]
        return df

    def predict(self, df):
        """
        Executa o pipeline completo: limpeza, feature engineering, preparação dos dados e retorna as predições em um DataFrame.
        
        Parâmetros:
        ----------
        df : pd.DataFrame
            DataFrame bruto de entrada.

        Retorno:
        -------
        df_predictions : pd.DataFrame
            DataFrame com os valores previstos de vendas para as amostras fornecidas.
        """
        # Limpa os dados
        df_clean = self.clean_data(df)
        # Engenharia de features
        df_feat = self.feature_engineering(df_clean)
        # Prepara os dados para o modelo
        df_ready = self.data_preparation(df_feat)
        # Faz predição
        preds = self.model.predict(df_ready)
        preds_exp = np.expm1(preds)
        df['prediction'] = preds_exp
        return df

```

### 13.3 API Handler

```python
import pandas as pd
from flask import Flask, request, Response
import os

app = Flask(__name__)

@app.route('/rossman/predict', methods=['POST'])
def rossman_predict():
    test_json = request.get_json()

    if test_json:

        if isinstance(test_json, dict):
            df = pd.DataFrame(test_json, index=[0]) 
        else:
            df = pd.DataFrame(test_json, columns=test_json[0].keys())

        pipeline = Rossman()

        df_predict = pipeline.predict(df)

        return Response(df_predict.to_json(orient='records'), status=200, mimetype='application/json')
        
    else: 
        return Response('{}', status=200, mimetype='application/json')


if __name__ == '__main__':
    port = int(os.environ.get("PORT", 5000))
    app.run('0.0.0.0', port=port)
```
